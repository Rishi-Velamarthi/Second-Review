# Distance vs Pga

```python
from math import log10
```

```python
def log(i):
    if i<0:
        return log10(-i)
    else:
        return log10(i)
```

```python
df['log(PGA)']=df['PGA'].apply(log)
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='Distance',y='log(PGA)',palette='mako',hue='Magnitude',alpha=0.5,edgecolor=None,data=df)
plt.xlabel('Distance (km)',size=15)
plt.ylabel('Log (PGA)',size=15)
```

```python
plt.figure(figsize=(12,2))
time=df['log(PGA)'].values

sns.histplot(time)
plt.xlabel('Log(PGA)')
```
